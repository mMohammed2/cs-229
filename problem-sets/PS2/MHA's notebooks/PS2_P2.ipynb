**Ans:**  
We have to show that the mean of all probs forecast by the model is equal to the fraction of ones in the original dataset.  

We know that the probability range (0, 1) covers all datapoints.

The intercept term translates the sigmoid curve, which originally intersects the y(or probability) axis at 0.5. 

We want to show that:

$$ \frac{1}{m} \sum_{i=1}^{m} \frac{1}{1+e^{-\theta^{T}x^{(i)}}} = \frac{1}{m} \sum_{i=1}^{m} \mathbf{I} \{ y^{(i)} = 1 \} $$

after the model using this function for predicting probabilities has converged to its optimal state.

### Likelihood

$$ \prod_{i=1}^{m} P(y^{(i)} \mid x^{(i)}) $$

### Log Likelihood

$$ \sum_{i=1}^{m} \log P(y^{(i)} \mid x^{(i)}) $$

Let $h(x^{(i)}) = \frac{1}{1+e^{-\theta^{T}x^{(i)}}}$

$$ L = \sum_{i=1}^{m} \left( \log (h(x^{(i)})) y^{(i)} + \log (1 - h(x^{(i)})) (1 - y^{(i)}) \right) $$

After training, we want the gradient to be zero for maximum likelihood.

### Sigmoid Function

The sigmoid function is defined as:

$$ \sigma(z) = \frac{1}{1 + e^{-z}} $$

To derive its derivative, we start by differentiating $\sigma(z)$ with respect to $z$:

$$ \frac{\partial \sigma}{\partial z} = \frac{\partial}{\partial z} \left( \frac{1}{1 + e^{-z}} \right) $$

Using the chain rule, we get:

$$ \frac{\partial \sigma}{\partial z} = \frac{1}{(1 + e^{-z})^2} \cdot \frac{\partial}{\partial z} (1 + e^{-z}) $$

$$ \frac{\partial \sigma}{\partial z} = \frac{1}{(1 + e^{-z})^2} \cdot (-e^{-z}) $$

Since $\sigma(z) = \frac{1}{1 + e^{-z}}$, we have $1 - \sigma(z) = \frac{e^{-z}}{1 + e^{-z}}$. Thus:

$$ \frac{\partial \sigma}{\partial z} = \sigma(z) (1 - \sigma(z)) $$


### Gradient of Log
$$ \frac{\partial \log(z)}{\partial x} = \frac{1}{z} \frac{\partial z}{\partial x} $$

Let $z = \theta^T x$, then

$$ \frac{\partial \sigma}{\partial z} = \sigma (1 - \sigma) $$

### Gradient of the Log Likelihood

$$ \frac{\partial L}{\partial \theta} = \frac{\partial L}{\partial h} . \frac{\partial h}{\partial \theta}$$

$$ \frac{\partial L}{\partial h} = \sum_{i=1}^{m} \left( \frac{y^{(i)}}{h(x^{(i)})} - \frac{1 - y^{(i)}}{1 - h(x^{(i)})} \right) $$

$$ \frac{\partial L}{\partial \theta} = \sum_{i=1}^{m} \left( \frac{y^{(i)}}{h(x^{(i)})} - \frac{1 - y^{(i)}}{1 - h(x^{(i)})} \right)\left( h(x^{(i)}) (1 - h(x^{(i)})) \right) x^{(i)} $$

$$ = \sum_{i=1}^{m} \left( y^{(i)} - h(x^{(i)}) \right) x^{(i)} $$

Since $\frac{\partial L}{\partial \theta} = 0$:

$$ \sum_{i=1}^{m} \left( y^{(i)} - h(x^{(i)}) \right) x_j^{(i)} = 0 $$

Let's set $x_j^{(i)} = 1$ because it will make evaluation easier. Since $x_j^{(i)} = 1$ for all $i$:

$$ \sum_{i=1}^{m} \left( y^{(i)} - h(x^{(i)}) \right) 1 = 0 $$

$$ \sum_{i=1}^{m} y^{(i)} = \sum_{i=1}^{m} h(x^{(i)}) $$

This is exactly what we set out to prove to begin with.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def plot_sigmoid_with_bias(X, slope, bias_values):
    plt.figure(figsize=(12, 8))
    for bias in bias_values:
        theta = np.array([bias, slope])
        z = np.dot(X, theta)
        y = sigmoid(z)
        plt.plot(X[:, 1], y, label=f'Bias = {bias}')
    
    plt.title('Sigmoid Curve with Different Bias Values')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.show()

# Generate input data
X = np.column_stack((np.ones(100), np.linspace(-10, 10, 100)))

# Set slope and bias values to test
slope = 0.7
bias_values = [-5, -2, 0, 2, 5]

# Plot sigmoid curves
plot_sigmoid_with_bias(X, slope, bias_values)
```
